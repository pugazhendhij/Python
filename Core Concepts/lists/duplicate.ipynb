✅ Method 1: Using a Set (Most Efficient, O(n))

````python
def findduplicate(arr):
    seen = set()
    duplicate = set()

    for i in arr:
        if i in seen:
            duplicate.add(i)
        else:
            seen.add(i)
    return list(duplicate)


print(findduplicate([2,3,4,2,3,1]))
````

# ✅ **How to Find Duplicate Numbers (Bullet Points)**

### **1️⃣ Create two sets**

* `seen` → keeps track of numbers we have already encountered
* `duplicate` → stores numbers that appear **more than once**

Sets are used because:

* Checking membership is **O(1)** (very fast)
* No duplicates are stored automatically

---

### **2️⃣ Loop through each number in the array**

---

### **3️⃣ Check if the number is already seen**

Condition:

```
if i in seen:
```

If TRUE:

* It means this number appeared **before**
* So this number is a **duplicate**

Action:

```
duplicate.add(i)
```

---

### **4️⃣ If the number is NOT seen yet**

Condition:

```
else:
    seen.add(i)
```

If TRUE:

* First time seeing this number
* Add it to `seen` set

---

### **5️⃣ After the loop ends**

* The `duplicate` set contains **all values that appeared more than once**
* Convert it to a list:

```
return list(duplicate)
```

How the Logic Works
- The core logic operates within the while loop, using simple arithmetic operations: 
1. last_digit = abs_n % 10: The modulo operator (%) gives the remainder of a division, which is the last digit of the number (e.g., 123 % 10 yields 3).
2. reversed_num = (reversed_num * 10) + last_digit: This line shifts the existing reversed number one decimal place to the left by multiplying by 10, and then adds the newly extracted last_digit to the end.
3. abs_n //= 10: The integer division operator (//) removes the last digit from the original number (e.g., 123 // 10 yields 12). 

The loop continues until the original number becomes 0, at which point all digits have been processed and the reversed_num is returned with the correct sign applied. Note that any trailing zeros in the original number will correctly be dropped as leading zeros in the integer result (e.g., 120 becomes 21). 

```python
def reverse_int(n):
    sign = -1 if n < 0 else 1
    abs_n = abs(n)
    reverse_num = 0

    while abs_n > 0:
        
        last_digit = abs_n % 10

        reverse_num = (reverse_num * 10) + last_digit

        abs_n = abs_n // 10

    return reverse_num * sign

print(reverse_int(9102)) 
```

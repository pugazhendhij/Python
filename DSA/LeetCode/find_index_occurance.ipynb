28. Find the Index of the First Occurrence in a String

Given two strings needle and haystack, return the index of the first occurrence of needle in haystack, or -1 if needle is not part of haystack.

 

Example 1:

Input: haystack = "sadbutsad", needle = "sad"
Output: 0
Explanation: "sad" occurs at index 0 and 6.
The first occurrence is at index 0, so we return 0.
Example 2:

Input: haystack = "leetcode", needle = "leeto"
Output: -1
Explanation: "leeto" did not occur in "leetcode", so we return -1.

````python
def find_index_occurance(haystack,needle):
    for i in range(len(haystack) - len(needle) + 1):
        window = haystack[i: i+ len(needle)]
        if window == needle:
            return i
    return -1

print(find_index_occurance("saibutsad","aib"))
````

# üß© Key Idea Behind This Solution

Instead of checking **all possible substrings**, we:

‚úÖ Check **only substrings of the same length as `needle`**
‚úÖ Slide this window **one character at a time**

This is called the **Sliding Window technique**.

# üîç Line-by-Line Explanation

---

## 1Ô∏è‚É£ Loop Range (MOST IMPORTANT LINE)

```python
for i in range(len(haystack) - len(needle) + 1):
```

### Why this range?

Let:

* `n = len(haystack)`
* `m = len(needle)`

You need **exactly `m` characters** to compare.

So the **last valid start index** is:

```
n - m
```

That‚Äôs why:

```
range(n - m + 1)
```

### Example

```text
haystack = "sadbutsad" (length = 9)
needle   = "sad"       (length = 3)

Last valid start index = 9 - 3 = 6
```

So indices checked:

```
0, 1, 2, 3, 4, 5, 6
```

---

## 2Ô∏è‚É£ Creating the Window

```python
window = haystack[i : i + len(needle)]
```

This extracts a **substring of the same length as needle**.

### Example when `i = 0`

```text
haystack[0:3] ‚Üí "sad"
```

### Example when `i = 1`

```text
haystack[1:4] ‚Üí "adb"
```

You are checking **only what is necessary**.

---

## 3Ô∏è‚É£ Comparison

```python
if window == needle:
    return i
```

* If the current window matches `needle`
* Return the **starting index**
* Because problem asks for the **first occurrence**, we stop immediately

---

## 4Ô∏è‚É£ If No Match Found

```python
return -1
```

* Loop finished
* No matching window
* So needle is not present

#### Missing Number from 1 to N
1. nums = [1,2,4,5] and return 3


````python
def missing_num(arr):
    n = len(arr) + 1
    total = n * (n + 1)  // 2
    return total - sum(arr)

print(missing_num([1,2,4,5]))

````



You are given a list of numbers from **1 to N**, but one number is missing.

Example:

```
[1, 2, 4, 5]
```

The numbers should be:

```
1, 2, 3, 4, 5
```

So the missing number is **3**.

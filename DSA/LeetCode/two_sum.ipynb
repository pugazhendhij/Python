Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

````python
def twosum(arr,target):
    seen = {}
    
    for i,num in enumerate(arr):
        complement = target - num
        
        if complement in seen:
            return [seen[complement],i]

        seen[num] = i

nums = [2,7,11,15]
target = 9
print(twosum(nums,target))
````

# âœ… **Purpose of this code**

This code is designed **only to find exactly TWO numbers** that add up to the target.
It works because:

* For every number, it checks if the *other* number already exists.
* If yes â†’ return the indices.
* If no â†’ store the current number for future matching.

This ensures exactly **one pair of numbers** is found.

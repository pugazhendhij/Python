35. Search Insert Position

Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.

 Example 1:

Input: nums = [1,3,5,6], target = 5
Output: 2

````python
def searchInsert(arr,target):
    arr.sort()

    start = 0
    end = len(arr) - 1
    
    while start <= end:
        mid = (start + end) // 2

        if arr[mid] == target:
            return mid
        
        elif target > arr[mid]:
            start = mid + 1
        else:
            end = mid - 1
        
    return start


print(searchInsert([1,3,5,6,7,8],9))
````

## 3Ô∏è‚É£ Key Observation (Very Important)

During binary search:

* Left side of `start` ‚Üí definitely smaller than target

When search ends:

> `start` is the ONLY safe place to insert

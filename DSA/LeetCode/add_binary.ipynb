67. Add Binary

Given two binary strings a and b, return their sum as a binary string.
 

Example 1:

Input: a = "11", b = "1"
Output: "100"
Example 2:

Input: a = "1010", b = "1011"
Output: "10101"

````python
def add_binary(a,b):
    i = len(a) - 1
    j = len(b) - 1
    carry = 0
    result = []

    while i >= 0 or j >= 0 or carry:

        total = carry
        if i >= 0:
            total += int(a[i])
            i -= 1

        if j >= 0:
            total += int(b[j])
            j -=1

        result.append(str( total % 2))
        carry = total // 2

    return "".join(reversed(result))

print(add_binary("1101","100"))
````

## üß† Core Idea (High Level)

This is **manual binary addition**, exactly how we add numbers on paper:

* Start from the **rightmost bit**
* Add bits **plus carry**
* Write the result bit
* Carry forward if needed

### 2Ô∏è‚É£ Carry Handling

* `carry` stores overflow from previous addition
* Ensures correct binary math

### 3Ô∏è‚É£ Total Calculation

```python

```

---

### 4Ô∏è‚É£ Extract Result Bit


* Gives the current binary digit (`0` or `1`)

### 5Ô∏è‚É£ Update Carry


* Determines if carry exists for next step

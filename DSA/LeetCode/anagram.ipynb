````python
def anagram_check(s1,s2):

    if len(s1) != len(s2):
        return False

    freq = {}
    for i in s1:
        freq[i] = freq.get(i,0) + 1

    for c in s2:
        if c not in freq or freq[c] == 0:
            return False
        freq[c] -= 1

    return True

print(anagram_check("apple","papel"))
````


# üöÄ **GOAL**

Check if `s1` and `s2` are **anagrams**
(meaning: same characters, same frequency, any order)

---

# üî• **STEP-BY-STEP EXPLANATION**

---

# **Step 1: Compare lengths**

```python
if len(s1) != len(s2):
    return False
```

`apple` ‚Üí length 5
`papel` ‚Üí length 5

Since both lengths match ‚Üí continue
If not equal ‚Üí immediately return **False**

---

# **Step 2: Create frequency dictionary for s1**

```python
freq = {}
for i in s1:
    freq[i] = freq.get(i, 0) + 1
```

Loop through `"apple"`:

| Character | Before | After |
| --------- | ------ | ----- |
| 'a'       | 0      | 1     |
| 'p'       | 0      | 1     |
| 'p'       | 1      | 2     |
| 'l'       | 0      | 1     |
| 'e'       | 0      | 1     |

Final `freq` dictionary:

```
{
  'a': 1,
  'p': 2,
  'l': 1,
  'e': 1
}
```

This tells us **how many times each character appears in s1**.

---

# **Step 3: Match characters from s2**

```python
for c in s2:
    if c not in freq or freq[c] == 0:
        return False
    freq[c] -= 1
```

We check `"papel"` one character at a time.

---

### ‚úî 1st character: `p`

* Is `'p'` in freq? ‚Üí Yes
* Is freq['p'] > 0? ‚Üí 2
* So reduce count:

```
freq['p'] = 1
```

---

### ‚úî 2nd character: `a`

* freq['a'] = 1 ‚Üí OK
* Reduce:

```
freq['a'] = 0
```

---

### ‚úî 3rd character: `p`

* freq['p'] = 1 ‚Üí OK
* Reduce:

```
freq['p'] = 0
```

---

### ‚úî 4th character: `e`

* freq['e'] = 1 ‚Üí OK
* Reduce:

```
freq['e'] = 0
```

---

### ‚úî 5th character: `l`

* freq['l'] = 1 ‚Üí OK
* Reduce:

```
freq['l'] = 0
```

---

# ‚ùó Why do we decrement?

Because every time we find a matching character in `s2`, we **use up one count** from `s1`.

If `s2` tries to use:

* A character not in s1 ‚Üí return False
* A character more times than s1 ‚Üí return False

This ensures exact frequency match.

---

# **Step 4: If loop completes ‚Üí it's an anagram**

Since no mismatch happened:

```python
return True
```

Output:

```
True
```

---

# ‚≠ê FINAL RESULT

`apple` and `papel` are **ANAGRAMS**
because they contain the same letters with the same frequency.
